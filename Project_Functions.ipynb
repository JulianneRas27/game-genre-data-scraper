```python
def game_genre_links(url):
    """ A function to get all game categories from a list of game categories found in girlsgogames.com
        Args: A string url to the all catgories link on girlsgogames.com
        Output: A list of urls leading to the different game categories 
    """

    # MAKE A LIST OF ALL THE GENRE LINKS
    import requests
    from bs4 import BeautifulSoup
    
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')
    
    urls = []
    for link in soup.find_all('a'):
        urls.append(link.get('href'))
    #print(urls)
        
    # REMOVE UNNECESSARY LINKS
    del urls[0:204]
    del urls[len(urls)-11:len(urls)]

    return urls

def game_links(all_category_urls):
    """
    A function to get game urls given a list of genre links.
    Args: A list of urls to different game genre webpages
    Output: A string representing the .txt file of the game urls separated by commas
    """

    # LOOP THROUGH ALL GENRES AND GET A LIST OF ALL GAME LINKS (PAGE 1)
    import requests
    from bs4 import BeautifulSoup

    each_list_of_games = []
    for genre in all_category_urls:
        reqs_g = requests.get(genre)
        soup_g = BeautifulSoup(reqs_g.text, 'html.parser')
        
        urls_g = []
        for link_g in soup_g.find_all('a'):
            urls_g.append(link_g.get('href'))

        #REMOVE UNNECESSARY LINKS
        #index_of_first = urls_g.index("https://www.girlsgogames.com/games")
        #del urls_g[0:index_of_first + 1]
        del urls_g[0:206]
        genre_name = genre[35:]
        link = f"https://www.girlsgogames.com/games/{genre_name}"
        if link in urls_g:
            index_of_second = urls_g.index(f"https://www.girlsgogames.com/games/{genre_name}")
            to_use = len(urls_g) - index_of_second
            del urls_g[len(urls_g)-to_use:len(urls_g)]
            for game in urls_g:
                each_list_of_games.append(game)
        else:
            continue

    each_list_of_games = ",".join(each_list_of_games)
    f = open("all_games.txt", "w")
    f.write(each_list_of_games)
    f.close()

    return "all_games.txt"

def get_game_titles(all_game_urls, file_name):
    """
    A function to get all game titles represented by given urls to girlsgogames.com games
    Args: A string representing a .txt file of urls separated by commas
    Output: A string representing a .txt file of all game titles separated by spaces
    """
    f = open(all_game_urls, "r")
    all_games = f.read()
    f.close()
    all_games = all_games.split(",")
    all_game_titles = []

    cropped_games = []

    for game in all_games:
        if game in cropped_games:
            continue
        else:
            cropped_games.append(game)

    for game in cropped_games:
        all_game_titles.append(game[34:len(game)].replace("-", " "))


    all_game_titles = " ".join(all_game_titles)


    g = open(file_name, "w")
    g.write(all_game_titles)
    g.close()

    return file_name


def boy_game_urls(url_to_use = "https://www.girlsgogames.com/games/boy-games"):
    """
    A function to get all of the game urls from the boy-game genre in girlsgogames.com
    Output: A list of urls representing all the games in the boys-games urls
    """
    import requests
    from bs4 import BeautifulSoup
    genre = url_to_use
    reqs_g = requests.get(genre)
    soup_g = BeautifulSoup(reqs_g.text, 'html.parser')
    each_list_of_games= []
    urls_g = []
    for link_g in soup_g.find_all('a'):
        game_link = link_g.get("href")
        urls_g.append(game_link)

    #REMOVE UNNECESSARY LINKS
    #index_of_first = urls_g.index("https://www.girlsgogames.com/games")
    #del urls_g[0:index_of_first + 1]
    del urls_g[0:206]

    index_of_second = urls_g.index(f"https://www.girlsgogames.com/games/boy-games")
    to_use = len(urls_g) - index_of_second - 1
    del urls_g[len(urls_g)-to_use:len(urls_g)]
    for game in urls_g:
        each_list_of_games.append(game)

    for i in range(1,9):
        page_game = other_pages(i)
        for game in page_game:
            each_list_of_games.append(game)


    boy_games = each_list_of_games
    g = open("boy_game_urls.txt", "w")
    g.write(",".join(boy_games))
    g.close()
    return "boy_game_urls.txt"


def other_pages(page_num):
    import requests
    from bs4 import BeautifulSoup
    genre = "https://www.girlsgogames.com/games/boy-games?page="+str(page_num)
    reqs_g = requests.get(genre)
    soup_g = BeautifulSoup(reqs_g.text, 'html.parser')
    each_list_of_games= []
    urls_g = []
    for link_g in soup_g.find_all('a'):
        game_link = link_g.get("href")
        urls_g.append(game_link)

    #REMOVE UNNECESSARY LINKS
    #index_of_first = urls_g.index("https://www.girlsgogames.com/games")
    #del urls_g[0:index_of_first + 1]
    del urls_g[0:206]

    index_of_second = urls_g.index(f"https://www.girlsgogames.com/games/boy-games")
    to_use = len(urls_g) - index_of_second - 1
    del urls_g[len(urls_g)-to_use:len(urls_g)]
    for game in urls_g:
        each_list_of_games.append(game)
    return each_list_of_games

def game_genres_v2(game_category_link, file_name):
    """
    A function to make a list of game genre urls found in every game
    Arg: A .txt file with game urls separated by commas
    Output: A list of the urls corresponding to the discovered game genres in a .txt file
    """

    import requests
    from bs4 import BeautifulSoup
    
    
    f = open(game_category_link, "r")
    all_games = f.read()
    f.close()
    all_games = all_games.split(",")
    each_list_of_labels = []
    f = open(file_name, "a")
    for game in all_games:
        reqs_l = requests.get(game)
        soup_l = BeautifulSoup(reqs_l.text, 'html.parser')
            
        urls_l = []
        for link_l in soup_l.find_all('a'):
                urls_l.append(link_l.get('href'))

        #remove unnecessary links
        pre_link = "https://www.girlsgogames.com/"
        if pre_link in urls_l:
            urls_l.remove("https://www.girlsgogames.com/")
        else:
            continue
        link_l = "https://www.girlsgogames.com/"
        link_l_two = 'https://www.girlsgogames.com/disclaimer'
        if link_l in urls_l and link_l_two in urls_l:
            key_index = urls_l.index("https://www.girlsgogames.com/")
            del urls_l[0:key_index+2]
            key_index_two = urls_l.index('https://www.girlsgogames.com/disclaimer')
            diff = len(urls_l) - key_index_two
            del urls_l[len(urls_l)-diff:len(urls_l)]
            each_list_of_labels.append(",".join(urls_l))
            urls = ",".join(urls_l)
            f.write(urls)
        
        else:
            continue
    f.close()
    print(each_list_of_labels)

def get_genre_labels(url_file, new_file_name):
    """
    A function to get the genres represented by the given links in a list.
    Input: a .txt file as a string with the urls separated by commas,
    a string representing the name of the new file created. 
    Output: a new .txt file as a string with the genres separated by spaces
    """

    f = open("boy_game_genre_urls.txt", "r")
    saved_labels = f.read()
    f.close()
    print(saved_labels)
    saved_labels = saved_labels.split(",")

    print(saved_labels)
    final_labels = []

    for genre in saved_labels:
        genre = genre[35:]
        final_labels.append(genre)

    print(final_labels)

    f = open(new_file_name,"a")

    for genre in final_labels:
        print(genre)
        genre = genre.replace("-", " ")
        genre = genre + " "
        f.write(genre)
        

    f.close()

def genre_dictionary(genre_list):
    """
    A function to turn a list of game genres into a dictionary where the keys are the genre
    as a string and the values are the number of instances of those genres. 
    Args: A list of genres separated by spaces in a .txt file
    Output: A dictionary variable
    """
    f = open(genre_list, "r")
    genre_list_string = f.read()
    f.close()
    genre_list_list = genre_list_string.split()
    label_dict = {}
    for label in genre_list_list:
        if label in label_dict:
            label_dict[label] = label_dict[label] + 1
        else:
            label_dict[label] = 1

    print(label_dict)
    return label_dict
```
