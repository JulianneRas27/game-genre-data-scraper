```python
import requests
f = open("all_games.txt", "r")
all_games = f.read()
f.close()
all_games = all_games.split(",")
all_game_titles = []

cropped_games = []

for game in all_games:
    if game in cropped_games:
        continue
    else:
        cropped_games.append(game)

for game in cropped_games:
    all_game_titles.append(game[34:len(all_games)].replace("-", " "))


all_game_titles = " ".join(all_game_titles)

print(all_game_titles)

g = open("all_game_titles.txt", "w")
g.write(all_game_titles)
g.close()
```

```python
import labellist
game_labels_all = []
for item in labellist.labels:
    for label in item:
        game_labels_all.append(label[35:len(label)])

label_dict = {}
for label in game_labels_all:
    if label in label_dict:
        label_dict[label] = label_dict[label] + 1
    else:
        label_dict[label] = 1

print(label_dict)

genre_instances = list(label_dict.values())
sorted_genre_instances = sorted(genre_instances)
print(sorted_genre_instances)
top_genre_instances = sorted_genre_instances[-21 : -1]
print(top_genre_instances)
top_genres = {}

for labels in label_dict:
    if label_dict[labels] in top_genre_instances:
        top_genres[labels] = label_dict[labels]

print(top_genres)

```
